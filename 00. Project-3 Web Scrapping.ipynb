```python
results = soup.find(id="ResultsContainer")
job_elements = results.find_all("div", class_="card-content")
i=0
jobs=[]
for results in job_elements:
    #Data from WEbpage to Variables
    job=results.find("h2", class_ = "title").text.strip() 
    com=results.find("h3").text.strip()
    loc=results.find("p").text.strip()
    date=results.find("time").text.strip()
    #Data from Variable to List
    jobs=jobs+[[job,com,loc,date]] #Saving All Details in list in list format
    i=i+1 #JObs COunt

#For Couting number of Jobs & Printing States specific Jobs
j=0
jobs_aa=[]
jobs_ae=[]
jobs_ap=[]
no_aa=0
no_ae=0
no_ap=0
while(j<i):
    state=jobs[j][2][-2:] #taking last 2 alphabets from Location for state
    state_job=jobs[j][0]
    if(state=="AA"):
        no_aa=no_aa+1 #Counting Jobs in AA
        jobs_aa=jobs_aa+[state_job] #Adding Jobs in AA to list
    elif(state=="AE"):
        no_ae=no_ae+1 #Counting Jobs in AE
        jobs_ae=jobs_ae+[state_job]#Adding Jobs in AE to list
    elif(state=="AP"):
        no_ap=no_ap+1 #Counting Jobs in AP
        jobs_ap=jobs_ap+[state_job] #Adding Jobs in AP to list
    j=j+1
print("Total Number of Jobs : ",i-1)
print("Number of Jobs in AA :",no_aa)
print(jobs_aa)
print()
print("Number of Jobs in AE :",no_ae)
print(jobs_ae)
print()
print("Number of Jobs in AP :",no_ap)
print(jobs_ap)
print()

```
